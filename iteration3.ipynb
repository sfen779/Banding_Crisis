```python
df['valueCrisis'] = df['tradevalue'] * df['contcrisis']
df.head()
df['valueCrisis'].plot.box()
```

```python
# change numeric to binary

df['contcrisis'] = df['contcrisis'].astype('category')
df['contcrisis'] = df['contcrisis'].cat.codes
df['contcrisis'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
# change numeric to categories:

bins = [-float('inf'), -0.5, 0.5, float('inf')]
lables = ['decrease', 'stable', 'increase']
df['cate_expg'] = pd.cut(df['expgrowth'], bins=bins, labels=lables)
df['cate_expg'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
num_cols = [cols for cols in df.columns if df[cols].dtype != 'O']
len(num_cols)
```

```python
SEP_DOLLAR = True
SEP_PERCENT = True

if SEP_DOLLAR:
    #For every column in df, if the column contains a $, make a new column with the value without the $
    for col in df.columns:
        if '$' in df[col].to_string():
            df[col + '_no_dollar'] = df[col].str.replace('$', '').str.replace(',', '')


if SEP_PERCENT:
    #For every column in df, if the column contains a %, make a new column with the value without the %
    for col in df.columns:
        if '%' in df[col].to_string():
            df[col + '_no_percent'] = df[col].str.replace('%', '').str.replace(',', '')
```

```python
#iteration:

from sklearn.utils import resample

df_high=df[df['developing']==0]

df_low=df[df['developing']==1]

df_low_boost=resample(df_low,replace=True,n_samples=len(df_high))

df_boosted=pd.concat([df_high,df_low_boost])

df_boosted['developing'].describe()

df_boosted['developing'].value_counts()

df_boosted['developing'].value_counts().plot.pie()

df_high_reduce=resample(df_high,replace=False,n_samples=len(df_low))

df_reduced=pd.concat([df_low,df_high_reduce])

df_reduced['developing'].describe()

df_reduced['developing'].value_counts()

df_reduced['developing'].value_counts().plot.pie()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler # For classification task
from sklearn.preprocessing import OrdinalEncoder
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
scale_cols = ['product','tradevalue']
df.columns
```

```python
cat_cols = [cols for cols in df.columns if df[cols].dtype == 'O']
cat_cols
```

```python
ct = ColumnTransformer([
    ('step1', StandardScaler(), scale_cols),
    ('step2', OrdinalEncoder(), cat_cols)
])

df.head()
```

```python
X = df.drop('tradeshare', axis=1)
Y = df['contcrisis']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
pipeline=Pipeline([
    ('cotf_step',ct),
    ('Random Forest',RandomForestClassifier())
    #('XGB',XGBClassifier())
])
pipeline.fit(X_train,Y_train)
```

```python

pipeline.score(X_train,Y_train)
```

```python
pipeline.score(X_test,Y_test)
```

```python
Y_pred = pipeline.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
print('Confusion matrix\n\n', cm)
print('\nTrue Positives(TP) = ', cm[0,0])
print('\nTrue Negatives(TN) = ', cm[1,1])
print('\nFalse Positives(FP) = ', cm[0,1])
print('\nFalse Negatives(FN) = ', cm[1,0])
```
